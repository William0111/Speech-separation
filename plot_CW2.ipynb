```python
# Set up the domain
l = np.linspace(-4,4,21) # 1D line
x,y = np.meshgrid(l,l)   # make a 2D mesh of points
z = x+1j*y               # 1j is python complex unit

# Function to plot complex potential flows
def plot(Phi,W,z=z):
    # Get real and imaginary parts
    x,y,psi,u,v = np.real(z),np.imag(z),np.imag(Phi),np.real(W),-np.imag(W)

    # Plot contours \psi
    plt.contour(x,y,psi,25)

    ## Plot quiver (vector field) of velocity
    plt.quiver(x,y,u,v)
    plt.axis('equal')
    plt.savefig('fig')
    plt.show()
```

```python
def circle_flow(a=2, zeta_0=0, z=z):
    # Define circle
    zeta_circ = a*np.exp(1j*np.linspace(0,2*np.pi))+zeta_0

    # Get zeta plane, masking out points near the singularity
    zeta = np.ma.array(z,mask=np.abs(z-zeta_0)<0.9*a)
  
    # Get potential and velocity
    Phi = zeta+a**2/(zeta-zeta_0); W = 1-a**2/(zeta-zeta_0)**2
    return zeta_circ,zeta,Phi,W

zeta_circ,zeta,Phi,W = circle_flow()
plt.figure(figsize=(3,3),dpi=200)
plt.plot(np.real(zeta_circ),np.imag(zeta_circ),'k',zorder=2)
plot(Phi,W,zeta)
```

```python
def joukowski_flow_plot(a, zeta_0):
    # get circle flow in zeta
    zeta_circ,zeta,Phi,W = circle_flow(3**0.5,zeta_0)
    
    # Map to z-plane
    z_ellipse = zeta_circ+1/zeta_circ
    z = zeta+1/zeta

    # Plot
    plt.figure(figsize=(3,3),dpi=200)
    plt.plot(np.real(z_ellipse),np.imag(z_ellipse),'k',zorder=2)
    plot1(Phi,W/(1-zeta**(-2)),z=z)
    print(np.abs(W/(1-zeta**(-2))).shape)
joukowski_flow_plot(a=3**0.5,zeta_0=0)
```

```python
def velocity_plot():
    def usoU(AR):
        theta=np.linspace(0,np.pi,1000)
        z=np.exp((-2.j)*theta)
        usoU=(1-z)/(1-z*(AR-1)/(AR+1))
        plt.plot(theta,usoU,label='AR={}'.format(AR))
        plt.legend(loc="upper right")
    

    usoU(2)
    usoU(4)
    usoU(8)
    plt.xlabel('i')
    plt.ylabel('us/U')
    plt.savefig('velocity_plot')
    #plt.show()
    
velocity_plot()

def pressure_c_plot():
    def cp(AR):
        theta=np.linspace(0,np.pi,1000)
        z=np.exp((-2.j)*theta)
        usoU=(1-z)/(1-z*(AR-1)/(AR+1))
        cp=1-abs(usoU)**2
        plt.plot(theta,cp,label='AR={}'.format(AR))
        plt.legend(loc="lower right")
        
    cp(2)
    cp(4)
    cp(8)
    plt.xlabel('theta')
    plt.ylabel('cp&Ue')
    plt.savefig('pressure_c_plot',dpi=300)
    plt.show()
    
pressure_c_plot()
```
